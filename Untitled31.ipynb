```python
import numpy as np
import tensorflow_datasets as tfds

# Load EMNIST letters dataset
(train_ds, test_ds), ds_info = tfds.load(
    'emnist/letters',
    split=['train', 'test'],
    as_supervised=True,
    with_info=True
)

# Convert train dataset
for images, labels in train_ds.batch(124800).take(1):
    X_train_letter = images.numpy()
    y_train_letter = labels.numpy()

# Convert test dataset
for images, labels in test_ds.batch(20800).take(1):
    X_test_letter = images.numpy()
    y_test_letter = labels.numpy()

print("Training:", X_train_letter.shape)
print("Testing:", X_test_letter.shape)
```

```python


from tensorflow.keras.utils import to_categorical

y_train_letter = to_categorical(y_train_letter, num_classes=27)
y_test_letter = to_categorical(y_test_letter, num_classes=27)
```

```python
import tensorflow as tf

# Creating base neural network
model = tf.keras.Sequential([

    # First Hidden Layer
    tf.keras.layers.Dense(256, activation='relu', input_shape=(784,)),

    # Second Hidden Layer
    tf.keras.layers.Dense(64, activation='relu'),

    # Third Hidden Layer
    tf.keras.layers.Dense(64, activation='relu'),

    # Output Layer for A–Z
    tf.keras.layers.Dense(27, activation='softmax')

])
```

| Layer          | Neurons | Purpose                     |
| -------------- | ------- | --------------------------- |
| Input Layer    | 784     | Receives 28×28 pixel values |
| Hidden Layer 1 | 256     | Learns important patterns   |
| Hidden Layer 2 | 64      | Extracts deeper features    |
| Hidden Layer 3 | 64      | Improves learning           |
| Output Layer   | 27      | Predicts alphabet classes   |


========

| Activation | Use                              |
| ---------- | -------------------------------- |
| `sigmoid`  | Mostly binary classification     |
| `softmax`  | Multi-class classification (A–Z) |

===========

Parameter calculation:

First hidden layer:

**Parameters = (Input x Output) + Bias**

= (784 x 256) + 256
= 200,704 + 256
= 200,960

==============

Second hidden layer:

= (256 x 64) + 64
= 16,384 + 64
= 16,448

==============

Third hidden layer:

= (64 x 64) + 64
= 4,096 + 64
= 4,160

============

Output layer:

= (64 x 27) + 27
= 1,728 + 27
= 1,755

Total:

Total parameters = 223,323
Trainable parameters = 223,323
Non-trainable parameters = 0

===========

**Flow:**

784 Pixels
      ↓
256 Neurons
      ↓
64 Neurons
      ↓
64 Neurons
      ↓
27 Output Classes (A-Z)

**This is essentially how ANN transforms pixel values into alphabet predictions.****

```python
# Compiling the model
model.compile(
    loss='categorical_crossentropy',   # Loss Function
    optimizer='adam',                  # Optimizer
    metrics=['accuracy']               # Evaluation Metric
)
```

| Parameter                         | Meaning           | Use                                                    |
| --------------------------------- | ----------------- | ------------------------------------------------------ |
| `loss='categorical_crossentropy'` | Loss function     | Calculates prediction error for multiple classes (A–Z) |
| `optimizer='adam'`                | Optimizer         | Updates weights to improve learning                    |
| `metrics=['accuracy']`            | Evaluation metric | Measures prediction accuracy                           |
| `model.layers[-1]`                | Last layer        | Displays output layer details                          |


```python
# Training the model

history = model.fit(
    X_train_letter,
    y_train_letter,
    batch_size=100,
    epochs=10,
    validation_data=(X_test_letter, y_test_letter)
)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Convert one-hot encoded labels back to normal labels
y_test_letter_eval = np.argmax(y_test_letter, axis=1)

# Predict test images
y_predicts = model.predict(X_test_letter)
y_predicts = np.argmax(y_predicts, axis=1)

# Confusion Matrix
con_mat = confusion_matrix(y_test_letter_eval, y_predicts)

# Alphabet labels
labels = [chr(i+64) for i in range(1,27)]

plt.figure(figsize=(15,15))

sns.heatmap(
    con_mat[1:,1:],     # ignore class 0
    annot=True,
    fmt='d',
    cmap='gray',
    xticklabels=labels,
    yticklabels=labels
)

plt.title(
    "Confusion Matrix for Alphabet Recognition",
    fontweight='bold',
    fontsize=15
)

plt.xlabel("Predicted Alphabet")
plt.ylabel("Actual Alphabet")

plt.show()
```

| Code                    | Meaning                             |
| ----------------------- | ----------------------------------- |
| `np.argmax()`           | Converts probabilities into labels  |
| `model.predict()`       | Predicts A–Z                        |
| `confusion_matrix()`    | Compares actual vs predicted labels |
| `labels=[chr(i+64)...]` | Creates A–Z labels                  |
| `annot=True`            | Displays numbers inside boxes       |
| `con_mat[1:,1:]`        | Removes unused class `0`            |
=========

Rows    → Actual alphabets,
Columns → Predicted alphabets,
Diagonal values → Correct predictions,
Other values → Wrong predictions,

========


```python
from sklearn.metrics import classification_report
import numpy as np

# Convert one-hot encoded labels back to normal labels
y_test_letter_eval = np.argmax(y_test_letter, axis=1)

# Predict test images
y_predicts = model.predict(X_test_letter)
y_predicts = np.argmax(y_predicts, axis=1)

# Alphabet labels A–Z
alphabet_labels = [chr(i+64) for i in range(1,27)]

# Generate report
print(
    classification_report(
        y_test_letter_eval,
        y_predicts,
        labels=range(1,27),
        target_names=alphabet_labels
    )
)
```
